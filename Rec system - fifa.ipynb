# Imports & Loading dataset 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('rec.csv')
df.dropna(inplace=True)
df.head()
```

# Creating Pivot Table

```python
pivot=df.pivot_table(columns='Name',values=[' Composure', ' Strength', 'Reactions', ' ShortPassing',
       ' HeadingAccuracy', ' BallControl'])
pivot
```

# Finding correlated players in the data

```python
#Selecting a random player
abang=pivot['A. Abang']
```

```python
abang.head()
```

```python
similar_to_abang=pivot.corrwith(abang)
```

```python
corr_abang=pd.DataFrame(similar_to_abang,columns=['corr'])
```

```python
corr_abang.head()
corr_abang.sort_values('corr',ascending=False).head(30)
```
